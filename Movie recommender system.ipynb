```python
# We have to preprocess above list of dictionary to ['Action','Adventure','Fantasy','Science Fiction']
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l    
```

```python
# obj=movie.iloc[0].genres
# convert(obj)
# To remove the error we have to convert string of list to list
```

```python
import ast
# ast.literal_eval()
```

```python
# apply() function acts as a map() function in Python.
# It takes a function as an input and applies this function to an entire DataFrame.
movie['genres']=movie['genres'].apply(convert)
```

```python
movie['genres']
```

```python
movie['keywords']=movie['keywords'].apply(convert)
```

```python
movie.iloc[0].cast
```

```python
# Will take top three actor/actress name from cast column
def convert_top_three(obj):
    l=[]
    c=0
    for i in ast.literal_eval(obj):
        if c!=3:
            l.append(i['name'])
            c=c+1
        else:
            break
    return l 
```

```python
movie['cast']=movie['cast'].apply(convert_top_three)
```

```python
movie.head()
```

```python
movie['crew'][0]
```

```python
# From crew we can will take director name.
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l    
```

```python
movie['crew']=movie['crew'].apply(fetch_director)
```

```python
movie.head()
```

```python
# Here overview is string , we have to convert it into list
movie['overview']=movie['overview'].apply(lambda x: x.split())
```

```python
# We have to remove the spaces between the words for better construction of model
movie['genres']=movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie.head()
```

```python
movie['keywords']=movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie['cast']=movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie['crew']=movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie.head()
```

```python
movie['tag']=movie['overview'] + movie['genres'] + movie['keywords'] + movie['cast'] + movie['crew'] 
```

```python
movie.head()
```

```python
new_df=movie[['movie_id','title','tag']]
```

```python
new_df.head()
```

```python
# Covert tag column which is list to string
new_df['tag']=new_df['tag'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tag']=new_df['tag'].apply(lambda x: x.lower()) # Recommended to convert to lowercase
```

```python
new_df.head()
```

```python
# We will use text vectorization to convert tag column to vectors.
# The process of converting words into numbers are called Vectorization.
# The idea is to get some distinct features out of the text for the model to train on, by converting text to numerical vectors.
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english') # Will provide us with top 5000 words with most frequency excluding stop_words in english such as in,but,thr,for,from etc
```

```python
vectors=cv.fit_transform(new_df['tag']).toarray()
# CountVectorizer gives us sparse matrix but we want numpy array so we will convert it to array 
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
# from above output we can see the following words 'accept','accepted','accepts' and many more are there.
# We can apply stemming to it (Natural Language processing)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tag']=new_df['tag'].apply(stem)
```

```python
cv.get_feature_names()
```

```python
# For all vectors we have to find cosine similarity . We cannot use euclidian similarity as it is not reliable for larger dataset
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# We can use cosine distance also. cosine distance is inversly proportional to cosine similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0].shape
```

RECOMMENDER FUNCTION

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0] # Extarcting the index of current movie
    distances=similarity[movie_index] # Getting the simirality score of current movie with all the other movie
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6] # Generating a list containing movies similar to current movie
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
